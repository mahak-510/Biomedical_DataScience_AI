Assuming again a binomial distribution, consider different pairs of ùëõ and ùëù:

 |ùëõ = 25| |ùëù = 0.3|

 |ùëõ = 50| |ùëù = 0.15|

 |ùëõ = 1000| |ùëù = 0.0075|
 
In each case, choose a suitable ùúÜ for the Poisson distribution and compare the probability mass functions of the binomial and the Poisson distribution on the interval [0, 16]. Under which conditions does the Poisson distribution closely approximate the binomial distribution?

 | n | p | $\lambda = np$|
 |:-----------|:----------:|-----------:
|25|0.3000|7.5|
|50|0.1500|7.5|
100|0.0750|7.5|
1000|0.0075|7.5|


```python
import math
import pandas as pd


diff_pairs: list[tuple[int, float]] = [
    (25,   0.30),
    (50,   0.15),
    (100,  0.075),
    (1000, 0.0075)
]

result=[]
for n, p in diff_pairs:
    lamda = n * p
    for k in range(17):  
        
        binomial_distr  = math.comb(n, k) * p**k * (1-p)**(n-k)
        poisson_distr = math.exp(-lamda) * lamda**k / math.factorial(k)
        result.append({
            "n": n, "p": p, "Œª": lamda, "k": k,
            "Binomial": binomial_distr,
            "Poisson":  poisson_distr,
            "difference ":   abs(binomial_distr - poisson_distr),
        })


df = pd.DataFrame(result) 
#printing all the rows of the dataframe
print(df)

```

**When P wasn't that much small there was a few percent difference between Binomial and Poisson distribution but as P get smaller and n grows higher than the difference become negligible. So the condition under which both distribution overlaps is when P is smaller and n is larger**
