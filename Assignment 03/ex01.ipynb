Part a: 21 blood units of blood group 0 (Rh+ and Rh-)? (1 point)

Blood group 0 corresponds to:

O+ → 35%

O- → 2%

So, total probability of blood group 0 = 35% + 2% = 37%.

We can model this as a binomial distribution:

Let 

X∼Binomial(n=42,p=0.37).

We are asked for the probability that 
X=21.

The binomial probability formula:
$$P(X = k) = \binom{n}{k}\,p^k\,(1 - p)^{n - k}$$

 
Thus:

$$P(X= 21) = \binom{42}{21}\,(0.37)^{21}\,(0.63)^{21}$$
 
Using calculator to solve:
=0.0288$$

Final answer:
Probability ≈ 0.0288

Part b: Probability of 6 blood units typed as Rh- 


 **Step 1: Identify parameters**

- \( n = 42 \)  
- \( k = 6 \)  
- \( p = 0.06 \) (probability of Rh-)
- \( q = 1 - p = 0.94 \)



**Step 2: Use the Binomial Probability Formula**

$$P(X = k) = \binom{n}{k} \cdot p^k \cdot q^{n-k}$$

$$P(X = 6) = \binom{42}{6} \cdot (0.06)^6 \cdot (0.94)^{36}$$

 **Step 3: Calculate each part (using calculator)**

#### 1. Binomial coefficient:
$$binom{42}{6} = \frac{42!}{6!(42 - 6)!} = 5245786$$

#### 2. Powers:
$$(0.06)^6 = 4.66 \times 10^{-9}$$

$$(0.94)^{36} \approx 0.115$$



 **Step 4: Multiply all parts**

$$P(X = 6) = 5245786 \cdot (4.66 \times 10^{-9}) \cdot 0.115$$

$$
\approx 5245786 \cdot 5.36 \times 10^{-10}
\approx 0.0527$$



###  Final Answer:

$$P(X = 6) \approx \boxed{0.0527}$$

So, there's approximately a **5.27%** chance that exactly 6 out of 42 donors are Rh⁻.



2. Of the 42 donated blood units, blood groups are observed with the following
frequencies: A+: 16, 0+: 14, B+: 5, AB+: 1, AB-: 2, B-: 3, 0-: 1, A-: 0. What is
the probability of gathering 42 blood units with exactly this composition of
blood groups? (2 points)



### 2. 
**Given observed frequencies:**

| Blood Group | Frequency |
| :--- | :--- |
| A+ | 16 |
| O+ | 14 |
| B+ | 5 |
| AB+ | 1 |
| AB- | 2 |
| B- | 3 |
| O- | 1 |
| A- | 0 |

The total is 42.

We can model this as a **multinomial probability**:

The multinomial formula is:


$$P = \frac{42!}{16!14!5!1!2!3!1!0!} (p_{A+})^{16}(p_{O+})^{14}(p_{B+})^5(p_{AB+})^1(p_{AB-})^2(p_{B-})^3(p_{O-})^1(p_{A-})^0$$

Where:
- \( p_{A+} = 0.37 \)
- \( p_{O+} = 0.35 \)
- \( p_{B+} = 0.09 \)
- \( p_{AB+} = 0.04 \)
- \( p_{AB-} = 0.06 \)
- \( p_{B-} = 0.06 \)
- \( p_{O-} = 0.02 \)
- \( p_{A-} = 0.01 \)

Substituting these values and simplifying gives the final probability or can be calculated using python as below:


```python
from scipy.stats import multinomial

# Total number of donors
n = 42

# Probabilities for each blood group (must sum to 1)
# Order: [O+, O-, A+, A-, B+, B-, AB+, AB-]
probs = [0.35, 0.02, 0.27, 0.02, 0.23, 0.01, 0.09, 0.01]

# Observed counts in the same order
counts = [17, 1, 10, 1, 9, 0, 3, 1]

# Calculate multinomial probability
probability = multinomial.pmf(counts, n, probs)

# Print result
print(f"Multinomial probability: {probability:.10f}")
```

### 3.
**Missing 7 blood donations.**

Given:
- On average, 5 people come in per hour.
- Want at least 7 people to come in one hour.

Assume Poisson distribution:


$$X \sim \text{Poisson}(\lambda = 5)$$


We want:

$$P(X \geq 7) = 1 - P(X \leq 6)$$

Use cumulative distribution for Poisson:

First calculate:

$$P(X \leq 6) = \sum_{k=0}^{6} \frac{e^{-\lambda} \lambda^k}{k!}$$

Where 
$$(\lambda = 5)$$



```python
from scipy.stats import poisson

# Mean number of donors per hour (λ)
lambda_val = 5

# We want: P(X ≥ 7) = 1 - P(X ≤ 6)
probability = 1 - poisson.cdf(6, mu=lambda_val)

# Display the result
print(" Poisson Probability ")
print(f"Mean number of donors per hour (λ): {lambda_val}")
print("We want: P(X ≥ 7)")
print(f"Probability that at least 7 people donate in the final hour: {probability:.4f}")
```
